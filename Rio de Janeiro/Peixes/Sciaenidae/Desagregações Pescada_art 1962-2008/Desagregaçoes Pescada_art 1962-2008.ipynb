```python
#Proporções das especies que receberao valores da desagregação
prop_Pesc_goete_art_1962_2008= np.load('/home/andre/Área de Trabalho/Reconstrução da Estatística Pesqueira BR/Rio de Janeiro/Peixes/Sciaenidae/Desagregações Pescada_art 1962-2008/prev_Pesc_goete_art.npy').flatten()
prop_Pesc_amarela_art_1962_2008= np.load('/home/andre/Área de Trabalho/Reconstrução da Estatística Pesqueira BR/Rio de Janeiro/Peixes/Sciaenidae/Desagregações Pescada_art 1962-2008/prev_Pesc_amarela_art.npy').flatten()
prop_Pesc_bicuda_art_1962_2008=np.load('/home/andre/Área de Trabalho/Reconstrução da Estatística Pesqueira BR/Rio de Janeiro/Peixes/Sciaenidae/Desagregações Pescada_art 1962-2008/prev_Pesc_bicuda_art.npy').flatten()
prop_Pesc_branca_art_1962_2008= np.load('/home/andre/Área de Trabalho/Reconstrução da Estatística Pesqueira BR/Rio de Janeiro/Peixes/Sciaenidae/Desagregações Pescada_art 1962-2008/prev_Pesc_branca_art.npy').flatten()
prop_Pesc_cambucu_art_1962_2008=np.load('/home/andre/Área de Trabalho/Reconstrução da Estatística Pesqueira BR/Rio de Janeiro/Peixes/Sciaenidae/Desagregações Pescada_art 1962-2008/prev_Pesc_cambucu_art.npy').flatten()
prop_Pesc_olhuda_art_1962_2008= np.load('/home/andre/Área de Trabalho/Reconstrução da Estatística Pesqueira BR/Rio de Janeiro/Peixes/Sciaenidae/Desagregações Pescada_art 1962-2008/prev_Pesc_olhuda_art.npy').flatten()

```

```python
#Soma das proporções
prop_soma = prop_Pesc_goete_art_1962_2008 + prop_Pesc_amarela_art_1962_2008 + prop_Pesc_bicuda_art_1962_2008  + prop_Pesc_branca_art_1962_2008 + prop_Pesc_cambucu_art_1962_2008 + prop_Pesc_olhuda_art_1962_2008

#Dividindo proporções pela soma para que de sempre 1
for i in range(len(prop_Pesc_goete_art_1962_2008)):
  prop_Pesc_goete_art_1962_2008[i] /= prop_soma[i] 
  prop_Pesc_amarela_art_1962_2008[i] /= prop_soma[i]
  prop_Pesc_bicuda_art_1962_2008 [i] /= prop_soma[i]
  prop_Pesc_branca_art_1962_2008[i] /= prop_soma[i]
  prop_Pesc_cambucu_art_1962_2008[i] /= prop_soma[i]
  prop_Pesc_olhuda_art_1962_2008[i] /= prop_soma[i]
```
